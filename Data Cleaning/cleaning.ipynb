```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Cleaning and enumuration of data

```python
dataCsv = pd.read_csv('data.csv')
dataCsv.head()
```

```python
dataCsv.info()
```

```python
dataCsv = dataCsv[dataCsv['baths'] != 0]                #remove rows with 0 baths
dataCsv = dataCsv[dataCsv['bedrooms'] != 0]             #remove rows with 0 bedrooms
dataCsv.to_csv('filteredData.csv', index=False)
dataCsv.head()
```

```python
#replacing strings with integers of price_bin
dataCsv = dataCsv.drop(['price_bin'], axis=1)
dataCsv.to_csv("filteredData.csv", index=False)
dataCsv.head()

```

```python
#removing area, area_marla column as they are not standard units

dataCsv = dataCsv.drop(['area','area_marla'], axis=1)
dataCsv.to_csv("filteredData.csv", index=False)
dataCsv.head()
```

```python
dataCsv = dataCsv[dataCsv['purpose'] == 'For Sale']   #remove rows with puprose other than sale
dataCsv = dataCsv.drop(['purpose'], axis=1)
dataCsv.to_csv("filteredData.csv", index=False)
dataCsv.head()
```

```python
dataCsv = dataCsv[dataCsv['property_type'] == 'House']   #remove rows with type other than house
dataCsv = dataCsv.drop(['property_type'], axis=1)
dataCsv.to_csv("filteredData.csv", index=False)
dataCsv.head()
```

```python
#replacing city names with integers to represnt cities
#Lahore = 0, Islamabad = 1, Karachi = 2, Rawalpindi = 3, Faisalabad = 4
dataCsv['city'] = dataCsv['city'].replace(['Lahore'], 0)
dataCsv['city'] = dataCsv['city'].replace(['Islamabad'], 1)
dataCsv['city'] = dataCsv['city'].replace(['Karachi'], 2)
dataCsv['city'] = dataCsv['city'].replace(['Rawalpindi'], 3)
dataCsv['city'] = dataCsv['city'].replace(['Faisalabad'], 4)

dataCsv.to_csv("filteredData.csv", index=False)
dataCsv.head()
```

```python
#replacing provice names with integers to represnt province
#Punjab = 0, Islamabad Capital = 1, Sindh = 2

dataCsv['province_name'] = dataCsv['province_name'].replace(['Punjab'], 0)
dataCsv['province_name'] = dataCsv['province_name'].replace(['Islamabad Capital'], 1)
dataCsv['province_name'] = dataCsv['province_name'].replace(['Sindh'], 2)

dataCsv.to_csv("filteredData.csv", index=False)
dataCsv.head()
```

```python
#removing locality column as same info is available in city, province_name, and location

dataCsv = dataCsv.drop(['locality'], axis=1)
dataCsv.to_csv("filteredData.csv", index=False)
dataCsv.head()
```

```python
#extract distinct values from location

uniqueVal = dataCsv['location'].unique()
print(uniqueVal)
# print(len(uniqueVal))

uniqueDict = dict()
for i in range(len(uniqueVal)):
    uniqueDict[uniqueVal[i]] = i

print(uniqueDict)
dataCsv = dataCsv.replace({'location': uniqueDict})
dataCsv.to_csv("filteredData.csv", index=False)
dataCsv.head()
```

Visualizing Data

```python
dataCsv.plot(kind='scatter', x='bedrooms', y='price')
plt.show()

dataCsv.plot(kind='scatter', x='baths', y='price')
plt.show()

dataCsv.plot(kind='scatter', x='location', y='price')
plt.show()

dataCsv.plot(kind='scatter', x='city', y='price')
plt.show()

dataCsv.plot(kind='scatter', x='province_name', y='price')
plt.show()

dataCsv.plot(kind='scatter', x='area_sqft', y='price')
plt.show()

dataCsv.plot(kind='scatter', x='latitude', y='price')
plt.show()

dataCsv.plot(kind='scatter', x='longitude', y='price')
plt.show()
```
